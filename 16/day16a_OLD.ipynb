```python
# try no. 3!
@dataclass
class Node:
    pt: Pt
    g: int
    h: int

    def __eq__(self, other: 'Pt | Node') -> bool:
        if isinstance(other, Node):
            return self.super().__eq__(other)
        if isinstance(other, Pt):
            return self.pt == other

@dataclass
class Vertex:
    pt: Pt
    edges: dict[Pt, list[Pt]]

    @classmethod
    def init(cls, pt, mat: StrMatrix) -> Self:
        return cls(pt=pt, edges={pt + br:[] for br in Pt.cardinals() if matget(mat, pt + br) != '#'})

def path_astar(maze: StrMatrix, start: Pt, end: Pt) -> list[Pt]:
    _frames: list[Image.Image] = []
    def snapshot(gif: bool=False):
        im = matimg(maze, colflt={
            (lambda p, v: v == '#'): 'black',
            (lambda p, v: p in seen): 'grey',
            (lambda p, v: p in final_path): 'blue',
            (lambda p, v: p == cursor): 'lime'
        }).resize((600,600), resample=0)
        if not gif:
            clear_output(wait=True)
            time.sleep(1)
            print([n.g + n.h for n in branches])
            display(im)
        if gif:
            _frames.append(im)
    
    cursor: Pt = start
    seen: set[Pt] = set()
    unchecked_verts: list[Vertex] = []
    final_path: list[Pt] = []

    def explore_until_vert(cursor: Pt, acc: list[Pt]) -> int | None:
        seen.add(cursor)
        while True:
            acc.append(cursor)
            snapshot(gif=True)
            if cursor == end:
                return 2

            branches: int = 0
            for direc in Pt.cardinals():
                future = cursor + direc
                if future in seen or matget(maze, future) == '#':
                    continue
                branches += 1
            
            if branches == 0:
                return 1
            if branches == 1:
                cursor = branches[0].pt
                seen.add(cursor)
                continue
            if branches > 1:
                # new_vt = Vertex(pt=cursor, edges={br:[] for br in branches})
                # if new_vt not in unchecked_verts:
                #     unchecked_verts.append(new_vt)
                return 0
            break
    try:
        cursor = start
        unchecked_verts.append(Vertex.init(cursor, maze))
        vt: Vertex | None = None
        while unchecked_verts:
            print('step')
            snapshot(gif=True)
            vt = unchecked_verts.pop()
            dead: set[Pt] = set()
            for br in vt.edges:
                ret = explore_until_vert(br, vt.edges[vt.pt + br])
                if ret == 2:
                    final_path.extend(vt.edges[br])
                    if _frames:
                        _frames[0].save('out.gif', save_all=True, append_images=_frames[1:], duration=100)
                    return final_path
                if ret == 1:
                    dead.add(br)
            print(vt.edges)
            new_vts = [
                Vertex.init(vt.edges[k][-1], maze) \
                for k in sorted(vt.edges, key=lambda i: (len(final_path) + len(vt.edges[i])) + i.distance(end)) \
                if k not in dead
            ]
            print(new_vts)
            unchecked_verts.extend(new_vts)
        if _frames:
            _frames[0].save('out.gif', save_all=True, append_images=_frames[1:], duration=100)
    except BaseException as e:
        import traceback
        print('ERR !!!')
        print('\n'.join(traceback.format_exception(e)))
        print(len(_frames))
        input()
        if _frames:
            _frames[0].save('out.gif', save_all=True, append_images=_frames[1:], duration=100)
```

```python
def heatmap(maze: Matrix, start: Pt, end: Pt) -> Image.Image:
    im = matimg(maze, colflt={(lambda p, v: v == '#'): 'black'})
    brush = ImageDraw.Draw(im)
    for pt, val in mat_iter(maze):
        if val != '#':
            g = pt.distance(start)
            h = pt.distance(end)
            brush.point((pt[1], pt[0]), f'hsv({120 * (g / start.distance(end))}, 100%, 100%)')
    return im
```
